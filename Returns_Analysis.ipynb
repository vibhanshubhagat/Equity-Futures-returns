```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate returns and statistics
def calculate_returns(data, period, price_col='SttlmPric'):
    data['Log Return'] = np.log(data[price_col] / data[price_col].shift(1))
    data = data.dropna(subset=['Log Return'])
    mean_return = data['Log Return'].mean() * 100
    max_return = data['Log Return'].max() * 100
    min_return = data['Log Return'].min() * 100
    std_dev_return = data['Log Return'].std() * 100
    statistics = {
        f'Mean {period} Return (%)': round(mean_return, 4),
        f'Max {period} Return (%)': round(max_return, 4),
        f'Min {period} Return (%)': round(min_return, 4),
        f'Standard Deviation {period} (%)': round(std_dev_return, 4)
    }
    return data, statistics

# Load the CSV files
current_file_path = '/content/TECHM_current.csv'
middle_file_path = '/content/TECHM_middle.csv'
far_file_path = '/content/TECHM_far.csv'
equity_file_path = '/content/TECHM.NS.daily.csv'
equity_weekly_file_path = '/content/TECHM.NS.weekly.csv'
equity_monthly_file_path = '/content/TECHM.NS.monthly.csv'

current_data = pd.read_csv(current_file_path)
middle_data = pd.read_csv(middle_file_path)
far_data = pd.read_csv(far_file_path)
equity_daily_data = pd.read_csv(equity_file_path)
equity_weekly_data = pd.read_csv(equity_weekly_file_path)
equity_monthly_data = pd.read_csv(equity_monthly_file_path)

# Ensure the data is sorted by date in ascending order
for data in [current_data, middle_data, far_data]:
    data['RptgDt'] = pd.to_datetime(data['RptgDt'])
    data.sort_values('RptgDt', inplace=True)

equity_daily_data['Date'] = pd.to_datetime(equity_daily_data['Date'])
equity_daily_data.sort_values('Date', inplace=True)

equity_weekly_data['Date'] = pd.to_datetime(equity_weekly_data['Date'])
equity_weekly_data.sort_values('Date', inplace=True)

equity_monthly_data['Date'] = pd.to_datetime(equity_monthly_data['Date'])
equity_monthly_data.sort_values('Date', inplace=True)

# Calculate daily returns and statistics
current_daily, current_daily_stats = calculate_returns(current_data.copy(), 'Daily')
middle_daily, middle_daily_stats = calculate_returns(middle_data.copy(), 'Daily')
far_daily, far_daily_stats = calculate_returns(far_data.copy(), 'Daily')

# Calculate weekly returns and statistics (taking every 7th day)
current_weekly, current_weekly_stats = calculate_returns(current_data.iloc[::7].copy(), 'Weekly')
middle_weekly, middle_weekly_stats = calculate_returns(middle_data.iloc[::7].copy(), 'Weekly')
far_weekly, far_weekly_stats = calculate_returns(far_data.iloc[::7].copy(), 'Weekly')

# Calculate monthly returns and statistics (taking every 30th day)
current_monthly, current_monthly_stats = calculate_returns(current_data.iloc[::30].copy(), 'Monthly')
middle_monthly, middle_monthly_stats = calculate_returns(middle_data.iloc[::30].copy(), 'Monthly')
far_monthly, far_monthly_stats = calculate_returns(far_data.iloc[::30].copy(), 'Monthly')

# Calculate equity returns
equity_daily_data['Log Return'] = np.log(equity_daily_data['Adj Close'] / equity_daily_data['Adj Close'].shift(1))
equity_daily_data = equity_daily_data.dropna(subset=['Log Return'])

equity_weekly_data['Log Return'] = np.log(equity_weekly_data['Adj Close'] / equity_weekly_data['Adj Close'].shift(1))
equity_weekly_data = equity_weekly_data.dropna(subset=['Log Return'])

equity_monthly_data['Log Return'] = np.log(equity_monthly_data['Adj Close'] / equity_monthly_data['Adj Close'].shift(1))
equity_monthly_data = equity_monthly_data.dropna(subset=['Log Return'])

# Plot the returns
plt.figure(figsize=(15, 12))

# Daily returns plot
plt.subplot(3, 1, 1)
plt.plot(current_daily['RptgDt'], current_daily['Log Return'] * 100, label='Current Month')
plt.plot(middle_daily['RptgDt'], middle_daily['Log Return'] * 100, label='Middle Month')
plt.plot(far_daily['RptgDt'], far_daily['Log Return'] * 100, label='Far Month')
plt.plot(equity_daily_data['Date'], equity_daily_data['Log Return'] * 100, label='Underlying Equity', linestyle='--', color='darkblue')
plt.xlabel('Date')
plt.ylabel('Daily Return (%)')
plt.title('Daily Current, Middle, Far Month Futures, and Underlying Equity Returns TECHM')
plt.legend()
plt.grid(True)

# Weekly returns plot
plt.subplot(3, 1, 2)
plt.plot(current_weekly['RptgDt'], current_weekly['Log Return'] * 100, label='Current Month')
plt.plot(middle_weekly['RptgDt'], middle_weekly['Log Return'] * 100, label='Middle Month')
plt.plot(far_weekly['RptgDt'], far_weekly['Log Return'] * 100, label='Far Month')
plt.plot(equity_weekly_data['Date'], equity_weekly_data['Log Return'] * 100, label='Underlying Equity', linestyle='--',color='darkblue')
plt.xlabel('Date')
plt.ylabel('Weekly Return (%)')
plt.title('Weekly Current, Middle, Far Month Futures, and Underlying Equity Returns TECHM')
plt.legend()
plt.grid(True)

# Monthly returns plot
plt.subplot(3, 1, 3)
plt.plot(current_monthly['RptgDt'], current_monthly['Log Return'] * 100, label='Current Month')
plt.plot(middle_monthly['RptgDt'], middle_monthly['Log Return'] * 100, label='Middle Month')
plt.plot(far_monthly['RptgDt'], far_monthly['Log Return'] * 100, label='Far Month')
plt.plot(equity_monthly_data['Date'], equity_monthly_data['Log Return'] * 100, label='Underlying Equity', linestyle='--',color='darkblue')
plt.xlabel('Date')
plt.ylabel('Monthly Return (%)')
plt.title('Monthly Current, Middle, Far Month Futures, and Underlying Equity Returns TECHM')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Creating dataframes for the statistics
daily_stats_df = pd.DataFrame([current_daily_stats, middle_daily_stats, far_daily_stats],
                              index=['Current Month', 'Middle Month', 'Far Month'])
weekly_stats_df = pd.DataFrame([current_weekly_stats, middle_weekly_stats, far_weekly_stats],
                               index=['Current Month', 'Middle Month', 'Far Month'])
monthly_stats_df = pd.DataFrame([current_monthly_stats, middle_monthly_stats, far_monthly_stats],
                                index=['Current Month', 'Middle Month', 'Far Month'])

# Displaying the statistics in tabulated form
print("Daily Sample Statistics:")
print(daily_stats_df)
print("\nWeekly Sample Statistics:")
print(weekly_stats_df)
print("\nMonthly Sample Statistics:")
print(monthly_stats_df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the CSV files
current_file_path = '/content/TECHM_current.csv'
middle_file_path = '/content/TECHM_middle.csv'
far_file_path = '/content/TECHM_far.csv'
equity_daily_file_path = '/content/TECHM.NS.daily.csv'

current_data = pd.read_csv(current_file_path)
middle_data = pd.read_csv(middle_file_path)
far_data = pd.read_csv(far_file_path)
equity_daily_data = pd.read_csv(equity_daily_file_path)

# Ensure the data is sorted by date in ascending order
current_data['RptgDt'] = pd.to_datetime(current_data['RptgDt'])
current_data.sort_values('RptgDt', inplace=True)

middle_data['RptgDt'] = pd.to_datetime(middle_data['RptgDt'])
middle_data.sort_values('RptgDt', inplace=True)

far_data['RptgDt'] = pd.to_datetime(far_data['RptgDt'])
far_data.sort_values('RptgDt', inplace=True)

equity_daily_data['Date'] = pd.to_datetime(equity_daily_data['Date'])
equity_daily_data.sort_values('Date', inplace=True)

# Merge the datasets on the dates
current_merged = pd.merge(current_data, equity_daily_data, left_on='RptgDt', right_on='Date', how='inner')
middle_merged = pd.merge(middle_data, equity_daily_data, left_on='RptgDt', right_on='Date', how='inner')
far_merged = pd.merge(far_data, equity_daily_data, left_on='RptgDt', right_on='Date', how='inner')

# Calculate the difference between the near month SttlmPric and the underlying equity Adj Close
current_merged['Difference'] = current_merged['SttlmPric'] - current_merged['Adj Close']
middle_merged['Difference'] = middle_merged['SttlmPric'] - middle_merged['Adj Close']
far_merged['Difference'] = far_merged['SttlmPric'] - far_merged['Adj Close']

# Count the number of contango and backwardation observations
current_contango = (current_merged['Difference'] > 0).sum()
current_backwardation = (current_merged['Difference'] < 0).sum()

middle_contango = (middle_merged['Difference'] > 0).sum()
middle_backwardation = (middle_merged['Difference'] < 0).sum()

far_contango = (far_merged['Difference'] > 0).sum()
far_backwardation = (far_merged['Difference'] < 0).sum()

# Plot the differences
plt.figure(figsize=(10, 5))

plt.plot(current_merged['RptgDt'], current_merged['Difference'], label='Current Month', color='blue')
plt.plot(middle_merged['RptgDt'], middle_merged['Difference'], label='Middle Month', color='green')
plt.plot(far_merged['RptgDt'], far_merged['Difference'], label='Far Month', color='orange')
plt.xlabel('Date')
plt.ylabel('Difference (SttlmPric - Adj Close)')
plt.title('Difference Between Futures SttlmPric and Underlying Equity Adj Close')
plt.legend()
plt.grid(True)
plt.show()

# Print the results
print("Number of Contango and Backwardation Observations:")
print(f"Current Month - Contango: {current_contango}, Backwardation: {current_backwardation}")
print(f"Middle Month - Contango: {middle_contango}, Backwardation: {middle_backwardation}")
print(f"Far Month - Contango: {far_contango}, Backwardation: {far_backwardation}")

# Displaying the first few rows of the merged data with the calculated difference
print("Current Month Differences:")
print(current_merged.head())
print("\nMiddle Month Differences:")
print(middle_merged.head())
print("\nFar Month Differences:")
print(far_merged.head())
```
